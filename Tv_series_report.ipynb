<h2 id="overview">üèÅ Overview</h2>

**_Formula 1: Drive to Survive_** is a British documentary series produced by Netflix in collaboration with Formula One. The series provides a behind-the-scenes look into the world of Formula 1 racing, including the lives of drivers, the dynamics within teams, and the drama of the racing season.

_First released in 2019, the show has aired seven seasons as of March 7, 2025._

**It has played a significant role in boosting global interest in Formula 1.**

A crossover event with **Full Swing** broadcast on Netflix in November 2023 called _The Netflix Cup_.

<h2 id="viewership-graph">üìà Viewership Graph</h2>

```python
import matplotlib.pyplot as plt

seasons = [1, 2, 3, 4, 5, 6, 7]
viewers = [3.5, 4.2, 5.0, 6.1, 6.7, 6.9, 7.2]

plt.plot(seasons, viewers, marker='o', color='blue')
plt.title('Viewership per Season')
plt.xlabel('Season')
plt.ylabel('Viewership (millions)')
plt.grid(True)
plt.show()
```

<h2 id="episodetoepisode-analysis">üîÑ Episode-to-Episode Analysis</h2>



```python
import numpy as np

changes = np.diff(viewers)

plt.bar(seasons[1:], changes, color='blue')
plt.title('Season-to-Season Change in Viewership')
plt.xlabel('Season')
plt.ylabel('Change in Viewership (millions)')
plt.axhline(0, color='black', linewidth=0.8)
plt.show()
```

<h2 id="observations">üßæ Observations</h2>

- The largest increase in viewership occurred between Seasons 3 and 4.
- After this peak, the growth in viewership from season to season has slowed down.
- This suggests that the series may have reached its prime audience, and future growth in viewership could be limited.
